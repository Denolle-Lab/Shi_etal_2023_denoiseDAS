```python
import numpy as np
from matplotlib import pyplot as plt

h = 75
thi = 15
src_trench_hor = h / np.tan(thi /180 * np.pi)
imsrc_trench_hor = src_trench_hor - 2*h*np.sin(2*thi /180 * np.pi)
x = np.arange(0, src_trench_hor, 1)
offset_src = src_trench_hor - x
offset_img = imsrc_trench_hor - x

Dp = np.sqrt(h**2 *4 + offset_img**2)
Ds = np.sqrt(h**2 + offset_src**2)

tp = Dp / 6.8
ts = Ds / 3.9


plt.figure(figsize=(12, 9))
plt.plot(x-150, tp, label='pPiP')
plt.plot(x-150, ts, label='S')
plt.ylim(0, 60)
plt.xlim(0, 80)
plt.legend()
```
